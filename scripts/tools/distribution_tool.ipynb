```python
import pandas as pd
from pathlib import Path

from util import distribution_summary, plot_distribution
```

```python
base_dir = Path.cwd().parent.parent
data_path = base_dir / "data" / "processed"

df = pd.read_csv(data_path / "combined_listings.csv", low_memory=False)
sold = pd.read_csv(data_path / "combined_sold.csv", low_memory=False)

print(df.columns.tolist())
```

# Functions
- `plot_distribution(data, column=None, ...)` — three-tier histogram (bulk / upper tail / extreme tail).
- `distribution_summary(data, column=None, ...)` — min, max, mean, median, and percentiles as a `pandas.Series`.

# Plot Distributions
- Panel 1: Bulk of Data
- Panel 2: Upper 1%
- Panel 3: Extreme cutoffs (can be set using `extreme_cutoff=` )

```python
plot_distribution(df, "OriginalListPrice");
```

```python
plot_distribution(df, "ListPrice");
```

```python
plot_distribution(sold, "ClosePrice");
```

### Any other numeric column


```python
plot_distribution(df, "LivingArea", extreme_cutoff=20_000);
```

# Distribution Summary
`distribution_summary` returns the numeric spread (count, min, max, mean, median, percentiles)

```python
COLS_TO_SUMMARIZE = [
    "ClosePrice",
    "LivingArea",
    "DaysOnMarket",
]

distribution_summary(df, COLS_TO_SUMMARIZE)
```
